```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset
try:
    df = pd.read_csv('googleplaystore.csv')
    print("Dataset loaded successfully!\n")
except FileNotFoundError:
    print("Error: 'googleplaystore.csv' not found. Please ensure the file is in the same folder.")

# 2. Data Cleaning (Essential for this specific dataset)
# Remove rows with missing Ratings
df.dropna(subset=['Rating'], inplace=True)

# Clean 'Installs' (Remove '+' and ',' then convert to float)
df['Installs'] = df['Installs'].str.replace('+', '', regex=False).str.replace(',', '', regex=False)
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')

# Clean 'Price' (Remove '$' and convert to float)
df['Price'] = df['Price'].str.replace('$', '', regex=False)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Drop any rows that became NaN during conversion
df.dropna(subset=['Installs', 'Price'], inplace=True)

# 3. Analysis: Grouping by Category
# We calculate the Mean Rating and the Count of apps per category
category_analysis = df.groupby('Category')['Rating'].agg(['mean', 'count']).sort_values(by='mean', ascending=False)

# 4. Professional Visualization
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")
plot = sns.barplot(x=category_analysis['mean'], y=category_analysis.index, palette='magma')

# Customizing the chart for better readability
plt.title('Average User Rating per App Category', fontsize=16, fontweight='bold')
plt.xlabel('Average Rating (Scale: 1-5)', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.xlim(3.8, 5.0)  # Zooming in to see the differences clearly
plt.tight_layout()

# Save the plot as an image for the assignment upload
plt.savefig('category_rating_output.png')
plt.show()

# 5. Final Text Output for Submission
print("-" * 30)
print("SUMMARY OF RESULTS")
print("-" * 30)
print(f"Total Apps Analyzed: {len(df)}")
print("\nTop 5 Highest Rated Categories:")
print(category_analysis['mean'].head(5))
print("\nBottom 5 Lowest Rated Categories:")
print(category_analysis['mean'].tail(5))

# OUTPUT
Dataset loaded successfully!

------------------------------
SUMMARY OF RESULTS
------------------------------
Total Apps Analyzed: 9360

Top 5 Highest Rated Categories:
Category
EVENTS                 4.435556
EDUCATION              4.389032
ART_AND_DESIGN         4.358065
BOOKS_AND_REFERENCE    4.346067
PERSONALIZATION        4.335987
Name: mean, dtype: float64

Bottom 5 Lowest Rated Categories:
Category
MAPS_AND_NAVIGATION    4.051613
VIDEO_PLAYERS          4.044595
TOOLS                  4.041512
ENTERTAINMENT          4.032143
DATING                 3.970769
Name: mean, dtype: float64
```
