```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset
# Ensure 'flight_price.csv' is in the same folder
try:
    df = pd.read_csv('flight_price.csv')
    print("Dataset loaded successfully!\n")
except FileNotFoundError:
    print("Error: 'flight_price.csv' not found. Please check the file path.")

# 2. Analysis Part A: Price vs. Days Left Until Departure
# We group by 'days_left' to see the average price trend
price_trend = df.groupby('days_left')['price'].mean().reset_index()

# 3. Analysis Part B: Airline Comparison (Delhi to Mumbai Route)
# We filter the data for the specific route requested
delhi_mumbai = df[(df['source_city'] == 'Delhi') & (df['destination_city'] == 'Mumbai')]
airline_comparison = delhi_mumbai.groupby('airline')['price'].mean().sort_values()

# 4. Visualization
sns.set_style("whitegrid")
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Days Left vs Price (Showing the Surge)
sns.lineplot(data=price_trend, x='days_left', y='price', ax=ax[0], color='red', linewidth=2)
ax[0].set_title('Price Surge vs. Days to Departure', fontsize=14, fontweight='bold')
ax[0].set_xlabel('Days Left Until Flight')
ax[0].set_ylabel('Average Ticket Price')
ax[0].invert_xaxis()  # Invert so 50 days is on left and 0 days is on right

# Plot 2: Airline Comparison (Delhi-Mumbai)
sns.barplot(x=airline_comparison.index, y=airline_comparison.values, ax=ax[1], palette='viridis')
ax[1].set_title('Average Price by Airline (Delhi-Mumbai)', fontsize=14, fontweight='bold')
ax[1].set_ylabel('Average Price')
ax[1].set_xlabel('Airline')
plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig('flight_analysis_output.png')
plt.show()

# 5. Final
print("-" * 45)
print("FLIGHT PRICE PREDICTION ANALYSIS REPORT")
print("-" * 45)
print(f"Total Records Analyzed: {len(df)}")
print("\n--- Price Surge Observation ---")
price_start = price_trend[price_trend['days_left'] == 50]['price'].values[0]
price_end = price_trend[price_trend['days_left'] == 1]['price'].values[0]
print(f"Avg Price 50 days out: {price_start:.2f}")
print(f"Avg Price 1 day out: {price_end:.2f}")
print(f"Total Surge: {((price_end - price_start) / price_start) * 100:.2f}% increase")

print("\n--- Delhi-Mumbai Airline Rankings (Cheapest to Premium) ---")
for airline, avg_price in airline_comparison.items():
    print(f"{airline:15} | Avg Price: {avg_price:.2f}")
print("-" * 45)
```
