```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset
try:
    df = pd.read_csv('hr_analytics.csv')
    print("Dataset loaded successfully!\n")
except FileNotFoundError:
    print("Error: 'hr_analytics.csv' not found.")

# 2. Data Cleaning
# Converting 'Attrition' from Yes/No to 1/0 for correlation analysis
if 'Attrition' in df.columns and df['Attrition'].dtype == 'object':
    df['Attrition_Num'] = df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

# 3. Analysis Part A: Correlation Heatmap
plt.figure(figsize=(12, 8))
# Selecting only numeric columns for correlation
numeric_df = df.select_dtypes(include=['float64', 'int64'])
correlation = numeric_df.corr()
sns.heatmap(correlation[['Attrition_Num']].sort_values(by='Attrition_Num', ascending=False),
            annot=True, cmap='RdYlGn', fmt='.2f')
plt.title('Key Drivers of Employee Attrition', fontsize=14)
plt.savefig('attrition_correlation.png')
plt.show()

# 4. Analysis Part B: Projects vs. Attrition
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='number_project', hue='Attrition', palette='Set1')
plt.title('Impact of Project Load on Attrition', fontsize=14)
plt.xlabel('Number of Projects')
plt.ylabel('Employee Count')
plt.savefig('projects_vs_attrition.png')
plt.show()

# 5. Statistical Summary
print("-" * 45)
print("HR ANALYTICS ATTRITION REPORT")
print("-" * 45)
if 'Attrition_Num' in df.columns:
    avg_attrition = df['Attrition_Num'].mean() * 100
    print(f"Overall Attrition Rate: {avg_attrition:.2f}%")

    print("\n--- Project Load Analysis ---")
    project_stats = df.groupby('number_project')['Attrition_Num'].mean() * 100
    for proj, rate in project_stats.items():
        print(f"Projects: {proj} | Attrition Rate: {rate:.2f}%")
print("-" * 45)
```
