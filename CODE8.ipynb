```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset
# Make sure the 'titanic.csv' file is in the same folder as this script
try:
    df = pd.read_csv('titanic.csv')
    print("Dataset loaded successfully!\n")
except FileNotFoundError:
    print("Error: 'titanic.csv' not found. Please check the file name.")

# 2. Data Cleaning
# We must fill missing Age values with the median so we don't lose data
df['Age'] = df['Age'].fillna(df['Age'].median())

# 3. Analysis Part A: Survival by Passenger Class (Pclass)
# Calculating the percentage of survivors in each class
pclass_survival = df.groupby('Pclass')['Survived'].mean() * 100

# 4. Analysis Part B: Survival by Age Group (Children vs. Adults)
# Defining 'Child' as Age < 18 and 'Adult' as Age >= 18
df['AgeGroup'] = df['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')
age_survival = df.groupby('AgeGroup')['Survived'].mean() * 100

# 5. Professional Visualization
sns.set_style("whitegrid")
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# Chart 1: Pclass vs Survival
sns.barplot(x=pclass_survival.index, y=pclass_survival.values, ax=ax[0], palette='Blues_d')
ax[0].set_title('Survival Rate by Passenger Class', fontsize=14, fontweight='bold')
ax[0].set_ylabel('Survival Percentage (%)')
ax[0].set_xlabel('Class (1 = Upper, 3 = Lower)')

# Chart 2: Age Group vs Survival
sns.barplot(x=age_survival.index, y=age_survival.values, ax=ax[1], palette='Oranges_d')
ax[1].set_title('Survival Rate: Children vs. Adults', fontsize=14, fontweight='bold')
ax[1].set_ylabel('Survival Percentage (%)')
ax[1].set_xlabel('Age Group')

plt.tight_layout()
plt.savefig('titanic_survival_report.png') # Saves the output as an image
plt.show()

# 6. Final
print("-" * 40)
print("TITANIC SURVIVAL ANALYSIS REPORT")
print("-" * 40)
print(f"Total Passengers Analyzed: {len(df)}")
print("\n--- Survival by Class ---")
for pclass, rate in pclass_survival.items():
    print(f"Class {pclass}: {rate:.2f}%")

print("\n--- Survival by Age Group ---")
for age_grp, rate in age_survival.items():
    print(f"{age_grp}: {rate:.2f}%")
print("-" * 40)
```
